```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms as transforms
```

```python
dataset_path = '/content/sfhq_dataset/a small sample (650 images)'
```

```python
def display_images(num_images=5):
    images = []
    image_files = os.listdir(dataset_path)

    # Load the specified number of images
    for i in range(min(num_images, len(image_files))):
        img_path = os.path.join(dataset_path, image_files[i])
        img = Image.open(img_path).convert("RGB")  # Open image and convert to RGB
        images.append(img)

    # Display the images
    plt.figure(figsize=(15, 5))
    for i, img in enumerate(images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
```

```python
display_images(num_images=5)
```

```python
from PIL import Image

# Function to get image dimensions
def get_image_dimensions(image_path):
    with Image.open(image_path) as img:  # Open the image file
        width, height = img.size  # Get dimensions
    return width, height

# Example usage
image_path = '/content/sfhq_dataset/a small sample (650 images)/SFHQ_pt2_00000002.jpg'
width, height = get_image_dimensions(image_path)
print(f"Width: {width}, Height: {height}")
```
