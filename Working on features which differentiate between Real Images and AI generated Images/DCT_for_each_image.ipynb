```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct
```

```python
def display_image(image_path, title="Image"):
    # Load the image
    image = cv2.imread(image_path)
    # Convert BGR (OpenCV default) to RGB for correct color display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.imshow(image)
    plt.axis('off')  # Hide axes
    plt.title(title)
    plt.show()
```

```python
# Load and preprocess the image
def load_grayscale_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
    image = cv2.resize(image, (256, 256))  # Resize for consistency
    return image

# Perform DCT and compute frequency magnitudes
def compute_dct_magnitudes(image):
    dct_transformed = dct(dct(image.T, norm='ortho').T, norm='ortho')  # Apply 2D DCT
    dct_magnitude = np.abs(dct_transformed)  # Take the magnitude
    return dct_magnitude
```

```python
def plot_frequency_randomness(image_path):
    image = load_grayscale_image(image_path)
    dct_magnitude = compute_dct_magnitudes(image)

    # Flatten the DCT magnitude for plotting
    dct_values = dct_magnitude.flatten()

    # Plot histogram of DCT values to show randomness
    plt.figure(figsize=(10, 6))
    plt.hist(dct_values, bins=50, color='blue', alpha=0.7)
    plt.yscale('log')  # Use log scale for better visibility of range
    plt.xlabel("DCT Magnitude (Frequency Strength)")
    plt.ylabel("Frequency (Log Scale)")
    plt.title("Frequency Domain Analysis - Randomness via DCT")
    plt.grid(True)
    plt.show()
```

```python
# Path to the input image
image_path = "/content/real 4.jpg"  # Replace with your image path

# Generate the randomness plot
plot_frequency_randomness(image_path)

display_image(image_path, title="Displayed Image")
```

```python
image_path = "/content/real 2.jpg"

plot_frequency_randomness(image_path)

display_image(image_path, title="Displayed Image")
```

**Spread of Frequencies**
*   A broader distribution of high frequencies (right side) often indicates more randomness and fine details.
*   A narrower distribution with low frequencies suggests less randomness, common in smooth or less textured images.
